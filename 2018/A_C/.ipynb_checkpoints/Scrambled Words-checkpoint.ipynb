```python
#T = test cases
#L = a list of L words->lowercase
#N = len(string)
T = int(input())
for t in range(T):
    L = int(input())
    dic_list = [s for s in input().split()]
    S1,S2,N,A,B,C,D = [s for s in input().split()]
    
    N = int(N)
    A = int(A)
    B = int(B)
    C = int(C)
    D = int(D)
    x1 = ord(S1)
    x2 = ord(S2)
    x = []
    x.append(x1)
    x.append(x2)
    S = []
    S.append(S1)
    S.append(S2)
    #自动生成测试所用的字符串
    for i in range(2,N):
        x.append((A*x[i-1] + B*x[i-2] + C) % D)
        S.append(chr(97+(x[i] % 26)))
print(S)      
```

```python
#计算列表中单词的频率
def cal_word_fre(word):
    num = {}
    for i in range(26):
        num[chr(i+97)] = 0
    for i in range(len(word)):
        if word[i] in num.keys():
            num[word[i]] = num[word[i]] + 1
    num_list = []
    num_list.append(word[0])
    num_list.append(word[1])
    num_list.extend(list(num.values()))
    return num_list
```

```python
#针对所生成的字符串进行枚举
def begin_test(dic_list,S_str):
    count = 0
    for i in range(len(dic_list)):
        for j in range(len(S_str)):
            list_test = cal_word_fre(S_str[j:j+len(dic_list[i])])
            print(list_test)
            print(cal_word_fre(dic_list[i]))
            if list_test == cal_word_fre(dic_list[i]):
                count += 1
                break
    return count
```

```python
begin_test(dic_list,S_str)
```
